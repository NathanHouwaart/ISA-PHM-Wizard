## ✅ SOLUTION: Factor Name Resolution

### The "N/A" Problem Explained

The ISA-JSON format uses **references** instead of embedding full objects. When the Python ISA tools serialize the data:

**What we expected:**
```json
{
  "category": {
    "factorName": "Time",
    "@id": "#study_factor/abc123"
  }
}
```

**What ISA-JSON actually produces:**
```json
{
  "category": {
    "@id": "#study_factor/abc123"
  }
}
```

The full factor details (like `factorName`) are stored separately in the `study.factors` array, and factor values reference them by `@id`.

### The Fix

To get the factor names, we need to:
1. Build a lookup map: `factor_id → factor_name`
2. When reading factor values, resolve the `@id` reference using this map

```python
# Create lookup
factor_id_to_name = {}
for factor in study['factors']:
    factor_id_to_name[factor['@id']] = factor['factorName']

# Resolve reference
factor_id = fv['category']['@id']
factor_name = factor_id_to_name[factor_id]
```

✅ Factor names now display correctly (Time, Cutting Speed, etc.)  
✅ Each run has correct per-run values (run_01, run_02, run_03...)  
✅ Table now shows actual values instead of N/A
